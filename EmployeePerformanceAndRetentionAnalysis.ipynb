```python
#Calculate the probability of an employee leaving based on factors like performance scores and department.
PerformanceScore_mean = df['PerformanceScore'].mean()
PerformanceScore_mean
```

```python
lowPerformance_attrition = df[df['PerformanceScore'] <= PerformanceScore_mean]['Attrition']
lowPerformance_attrition
```

```python
#the probability of an employee leaving based on performance scores
#Attrition Rate (Marginal Probability)
#prob_attrition = (df['Attrition'] == 1).mean()
#Conditional Probability
#P(Attrition=1Â |Â PerformanceÂ â‰¤Â PerformanceScore_mean)
p_cond = df[df['PerformanceScore'] <= PerformanceScore_mean]['Attrition'].mean()
print("P(Attrition==1 | Low Performance):", p_cond)
```

```python
#dataset of employees leaving based on sales department #'Sales':3 ,'HR':1, 'Engineering':0 ,'Marketing':2
print(df[(df['Department'] == 3) & (df['Attrition'] == 1)]) #14 rows in output
```

```python
#the probability of an employee leaving based on department
#Joint Probability
#Probability that employee is in Sales and left:
p_joint = len(df[(df['Department'] == 3) & (df['Attrition'] == 1)]) / len(df)
p_joint
```

```python
#Use Bayes' Theorem to find the probability of employee attrition given performance score.
prob_attrition_yes = (df['Attrition']==1).mean()
prob_cond = len(df[(df['PerformanceScore'] <= PerformanceScore_mean) & (df['Attrition'] == 1)])/len(df)
prob_low_performance = len(df[df['PerformanceScore'] <= PerformanceScore_mean])/len(df)
p_perf = (df['PerformanceScore']<= PerformanceScore_mean).mean()
Bayes_prob = (prob_attrition_yes * prob_cond)/prob_low_performance
print("P(Attrition==1 | Low Performance):", Bayes_prob)
print("P(Attrition==1):", prob_attrition_yes)
print("P(Low Performance):", prob_low_performance,p_perf)
print("P(Attrition==1 & Low Performance):", prob_cond)
```

```python
#Hypothesis Testing - Test whether the mean performance score differs across departments.
#This is a one-way ANOVA test (Analysis of Variance).
#Null Hypothesis (Hâ‚€): Mean performance score is the same across all departments.
#Alternative Hypothesis (Hâ‚): At least one department has a different mean performance score.
from scipy.stats import f_oneway
# Group data by department
#df.groupby('Department')
#This splits the DataFrame into smaller DataFrames, one for each unique Department.
#So if our Department column has HR, IT, Sales,engineering we get 4 groups.
#Each group is a mini DataFrame containing only rows of that department. Each name is the department name ('HR', 'IT', 'Sales','engineering').
#group['Performance Score']
#From each group, we select only the Performance Score column. That gives us a Pandas Series of performance scores for that department.
#.values converts that Series into a NumPy array. So instead of a Pandas object, we just have plain arrays of numbers (required by f_oneway in SciPy).
groups = [group['PerformanceScore'].values for name, group in df.groupby('Department')]
# Run ANOVA
f_stat, p_val = f_oneway(*groups)
print("F-statistic:", f_stat)
print("p-value:", p_val)
if p_val < 0.05:
    print("Reject Hâ‚€ â†’ Performance score differs across departments.")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference in performance scores.")
#p-value < 0.05 â†’ evidence that performance scores differ across at least one department.
#p-value â‰¥ 0.05 â†’ no statistically significant difference in mean performance across departments.
```

```python
#Deep Learning for employee Performance Prediction(Regression Model)
#Build a feedforward neural network: Using Keras Sequential API
#Input layer - Employee features like Age, Salary, Department etc.
#Hidden layers - Dense layers with activation functions (e.g., ReLU).
#Output layer - Predicted Performance Score.
#Train the model using Mean Squared Error as the loss function.
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
# Preparing the dataset. 
#min max scaled dataset
scaled_X = mms.fit_transform(X)
X_scaled = pd.DataFrame(scaled_X ,columns=['EmployeeID', 'Age', 'Department', 'Salary', 'YearsAtCompany',
       'Attrition'])
```

```python
#Test-train split dataset: X_train,X_test,Y_train,Y_test
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled, Y, test_size=0.33)
```

```python
#X_train.shape[1] gives the number of features (columns). Example:our dataset has shape (100, 6) â†’ X_train.shape[0] = 100 (rows / samples)
#X_train.shape[1] = 6 (features) But Keras requires a tuple to define the shape. If we write shape=6, Keras will think itâ€™s an integer, not a tuple.
#To make it a tuple with one element, we add a comma:
# feed forward NN - Sequential 
#Define the model
model = Sequential([
    Input(shape=(X_train.shape[1],)), #input layer 
    #Dense(100, activation='relu', input_shape=(X_train.shape[1],)), #we could also merge input layer and first hidden layer 
    Dense(100,activation='relu'), #hidden layer
    Dense(100,activation='relu'),
    Dense(1)#output layer, for regression we can skip activation function as we are predicting continuous value
])
```

```python
# Compile model (MSE for regression) 
model.compile(optimizer='adam', loss='mse', metrics=['mae']) #metrics could be mae,mse,huber loss (Huber combines the best of MSE and MAE)
#from tensorflow.keras.losses import Huber
#model.compile(optimizer='adam',loss=Huber(delta=1.0),metrics=['mae', 'mse']) # we can tune delta
#Start with MSE (fastest convergence).If we see outliers messing up results â†’ switch to Huber.Use MAE if interpretability is most important
#Use MSE â†’ when errors are normally distributed (no big outliers).
#Use MAE â†’ when data has many outliers, and we care about median-like behavior.
#Use Huber Loss â†’ when we want the stability of MSE and the robustness of MAE.Example: Salary prediction dataset where some employees have unusually high salaries.
```

```python
#Prepare model summary
model.summary()
```

```python
#training the model
training_logs = model.fit(
    X_train, Y_train,
    validation_split=0.2,  # some of training used for validation
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
#model evaluation on test data
test_loss, test_mae = model.evaluate(X_test, Y_test, verbose=0)
print(f"Test MSE: {test_loss:.4f}")
print(f"Test MAE: {test_mae:.4f}")
```

```python
# Predictions
Y_pred = model.predict(X_test)
```

```python
#model visualization
plt.plot(training_logs.history['loss'], label='Training Loss')
plt.plot(training_logs.history['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.33)
```

```python
attrition_model = Sequential([
    Input(shape=(x_train.shape[1],)),
    Dense(100,activation='relu'),
    Dense(50,activation='relu'),
    Dense(1,activation='sigmoid') #for binary classification
])
```

```python
attrition_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
#attrition_model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
# opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
attrition_model.summary()
```

```python
##early stopping:Stop training when a monitored metric has stopped improving.
import tensorflow as tf
early_Stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
#training the model
attrition_model_log = attrition_model.fit(
    x_train,y_train,
    validation_split =0.2,
    epochs=50, 
    batch_size=16,#When training a neural network,we donâ€™t usually update weights after each single sample (too slow) or after the entire dataset either (too memory heavy).Instead, we split the dataset into mini-batches of size = batch_size.
    verbose=1,
    callbacks=early_Stopping
)
```
