```python
%matplotlib inline
import C.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
# Plotting the line
plt.plot(x, y, marker='o', linestyle='-', color='blue')

# Add title
plt.title('Line Plot of X vs Y', fontdict={'fontsize':12, 'fontweight': 'bold', 'color': 'green'})

# Add axis labels
plt.xlabel('X-Axis', fontdict={'fontsize':10, 'color': 'red'})
plt.ylabel('Y-Axis', fontdict={'fontsize':10, 'color': 'red'})

# Add grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# Show plot
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students,marks,align='center',width=0.5,color='orange')
plt.xlabel('Students', fontdict={'fontsize':10, 'color': 'red'})
plt.ylabel('marks', fontdict={'fontsize':10, 'color': 'red'})
plt.ylim(0,100)
plt.title('Bar graph of students and their marks', fontdict={'fontsize':12, 'fontweight': 'bold', 'color': 'green'})
plt.show()
```
