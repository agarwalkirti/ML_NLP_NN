```python
corpus = """
Hello, Welcome to Kirti's journey! self learning data science course. 
Please do appreciate the hard work and dedication.

"""
```

```python
import nltk
nltk.download('punkt') 
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
### paragraph to sentence
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)
```

```python
### paragraph to words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
### paragraph to words
from nltk.tokenize import WordPunctTokenizer 
WPtokenizer = WordPunctTokenizer()
WPtokenizer.tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
TBWtokenizer = TreebankWordTokenizer()
TBWtokenizer.tokenize(corpus)
```

```python
from nltk.stem.porter import PorterStemmer
portstemming = PorterStemmer()
```

```python
### words = nltk.word_tokenize(i)
words=["eating","eats","eaten","writing","writes","programming","programs","history","finalized","fairly","sportingly","finally","finalize","congratulations","understanding","sitting"]
```

```python
for word in words:
    print(word + "---->"+portstemming.stem(word))
```

```python
### LANcaster stemming algorithm
from nltk.stem import LancasterStemmer
from nltk.corpus import stopwords
```

```python
lancaster = LancasterStemmer()
stopwords.words('english')
```

```python
for word in words:
    if word not in set(stopwords.words('english')):
        print(word + "---->"+lancaster.stem(word))
```

```python
### Regexp stemming algorithm
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('$ing$|s$|e$|able$',min=4)
```

```python
for word in words:
    print(word + "---->"+reg_stemmer.stem(word))
```

```python
###snowball stemmer
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english',ignore_stopwords=False)
```

```python
for word in words:
    print(word + "---->"+snowballstemmer.stem(word))
```

```python
##WordNet Lemmatizer
from nltk.stem import WordNetLemmatizer
WNlemmatizer = WordNetLemmatizer()
```

```python
'''
pos - 
Noun -n
verb -v
adjective -a
adverb - r
'''
for word in words:
    print(word + "---->"+WNlemmatizer.lemmatize(word,pos='v'))
```

```python
corpus =[ """
India, officially the Republic of India, is a country in South Asia.
It is the seventh-largest country by area; the most populous country since 2023;and, 
since its independence in 1947, the worlds most populous democracy.
Bounded by the Indian Ocean on the south, the Arabian Sea on the southwest,
and the Bay of Bengal on the southeast, it shares land borders with Pakistan to the west; 
China, Nepal, and Bhutan to the north; and Bangladesh and Myanmar to the east. In the Indian Ocean,
India is near Sri Lanka and the Maldives; its Andaman and Nicobar Islands share a maritime border 
with Myanmar, Thailand, and Indonesia.
Modern humans arrived on the Indian subcontinent from Africa no later than 55,000 years ago.
Their long occupation, predominantly in isolation as hunter-gatherers, has made the region highly diverse.
Settled life emerged on the subcontinent in the western margins of the Indus river basin 9,000 years ago,
evolving gradually into the Indus Valley Civilisation of the third millennium BCE.[30] By 1200 BCE, 
an archaic form of Sanskrit, an Indo-European language, had diffused into India from the northwest.
Its hymns recorded the early dawnings of Hinduism in India.
India pre-existing Dravidian languages were supplanted in the northern regions.
By 400 BCE, caste had emerged within Hinduism,and Buddhism and Jainism had arisen, 
proclaiming social orders unlinked to heredity.
Early political consolidations gave rise to the loose-knit Maurya and Gupta Empires.
Widespread creativity suffused this era, but the status of women declined,
and untouchability became an organised belief.In South India, 
the Middle kingdoms exported Dravidian language scripts and religious cultures 
to the kingdoms of Southeast Asia
"""]
```

```python
### bag of words (BOW)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer() ###for binary we can give binary=true in parameters
```

```python
###applying stopwords, lemmatize
import re
for i in range(len(corpus)):
    review = re.sub('[^a-zA-Z]', ' ' , corpus[i]) ##other than ato z and A to Z remove all special characters
    review = review.lower() # lower all letters
    review = review.split() ## split each and every sentence
    ##removing stopwords
    review = [WNlemmatizer.lemmatize(word,pos='v') for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
corpus
```

```python
X=cv.fit_transform(corpus)
###gives index, feature count
cv.vocabulary_
```

```python
#binary BOW
cv2 = CountVectorizer(binary=True) ##for binary we can give binary=true in parameters
X=cv2.fit_transform(corpus)
###gives dictionary contains word-index mapping, feature count
cv2.vocabulary_ 
X[0].toarray()
```

```python
###N-gram
cv3 = CountVectorizer(binary=True,ngram_range=(2,3))
X=cv3.fit_transform(corpus)
###gives index, feature count
cv3.vocabulary_
X[0].toarray()
```

```python
##TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
cv4 = TfidfVectorizer()
X=cv4.fit_transform(corpus)
corpus[0]
```

```python
###gives index, feature count
cv4.vocabulary_
```

```python
X[0].toarray()
```

```python
##TF-IDF & ngram, max features = top 3 max features
from sklearn.feature_extraction.text import TfidfVectorizer
cv5 = TfidfVectorizer(ngram_range=(1,3),max_features=3)
X=cv5.fit_transform(corpus)
###gives index, feature count
cv5.vocabulary_
X[0].toarray()
```
