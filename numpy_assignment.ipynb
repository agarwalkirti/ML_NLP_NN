```python
import numpy as np
array_np=np.array(range(1,21))
print(array_np)
```

```python
summ=np.sum(array_np)
meann= np.mean(array_np)
mediann =np.median(array_np)
sd=np.std(array_np)
print(f"sum {summ} mean {meann } median {mediann} sd {sd}")
indicess=np.where(array_np>10)
print(f"index in array whose value > 10 {indicess}")
```

```python
matrix_np=np.arange(1,17).reshape(4,4)
print(matrix_np)
```

```python
matrix_np=np.arange(1,17).reshape(4,4)
print(matrix_np)
matrix_transpose_np=matrix_np.transpose()
print(matrix_transpose_np)
```

```python
matrix_np_row_sum=np.sum(matrix_np,axis=1)
matrix_np_col_sum=np.sum(matrix_np,axis=0)
print("row wise sum:",matrix_np_row_sum)
print("col wise sum:",matrix_np_col_sum)
```

```python
import random
matrix_np_1=np.random.randint(1, 21, size=(3, 3))
matrix_np_2=np.random.randint(1,20, size=(3,3))
print(f"matrix 1: {matrix_np_1}, \n matrix 2: {matrix_np_2}")
matrix_np_sum=matrix_np_1+matrix_np_2
matrix_np_mul=matrix_np_1 * matrix_np_2
matrix_np_sub=matrix_np_1-matrix_np_2

print(f"element wise details:\n sum: {matrix_np_sum},\n multipication: {matrix_np_mul}, \nsubtraction: {matrix_np_sub}")
```

```python
matrix_reshape=np.arange(1,13).reshape(3,4)
print(matrix_reshape)
matrix_reshape_row_col_sliced=matrix_reshape[0:2,2:4]
print(f"sliced matrix with the first two rows and last two columns: {matrix_reshape_row_col_sliced} ")
```

```python
# Step 1: Create a 1D array of size 12
arr_1d = np.arange(1, 13)  # [1, 2, ..., 12]

# Step 2: Reshape to 3x4
arr_2d = arr_1d.reshape(3, 4)

# Step 3: Slice first two rows and last two columns
sliced = arr_2d[0:2, 2:4]

# Display
print("Original 1D Array:\n", arr_1d)
print("\nReshaped 2D Array (3x4):\n", arr_2d)
print("\nSliced Array (first two rows, last two columns):\n", sliced)
```
